```python
from pymatgen.ext.matproj import MPRester
from IPython.display import Image
from pymatgen.core.operations import SymmOp
from pprint import pprint
with MPRester("P2FTs1l6Xaon8AOI") as m: 
    structure = m.get_structure_by_material_id("mp-28240")
```

```python
pprint("Element de symétrie -x+y, -x, z")
pprint("Coordonées initiales : [0.0,  0.0,  0.8]")
S1 = SymmOp.from_xyz_string("-x+y, -x, z")
pprint("Matrice de rotation")
pprint(S1.rotation_matrix)
pprint("Vecteur de translation")
pprint(S1.translation_vector)
pprint("Coordonées finales : ")
pprint(S1.operate([0.0,  0.0,  0.8]))

Image(filename='-x+y, -x, z.png')
```

```python
pprint("Element de symétrie -y, -x, z")
pprint("Coordonées initiales : [0.0,  0.0,  0.8]")
S2 = SymmOp.from_xyz_string("-y, -x, z")
pprint("Matrice de rotation")
pprint(S2.rotation_matrix)
pprint("Vecteur de translation")
pprint(S2.translation_vector)
pprint("Coordonées finales : ")
pprint(S2.operate([0.0,  0.0,  0.8]))


Image(filename='-y, -x, z.png')
```
