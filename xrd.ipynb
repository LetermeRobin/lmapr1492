```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
from IPython.display import HTML, display
import tabulate 
with MPRester("P2FTs1l6Xaon8AOI") as m: 
    struct = m.get_structure_by_material_id("mp-28240")
    
xrdcalc = XRDCalculator(wavelength = 'CuKa', symprec = 0, debye_waller_factors = None)

xrdpattern = xrdcalc.get_pattern(structure = struct, scaled = True, two_theta_range = (0,90))
xrdcalc.show_plot(struct)

pic1 = [xrdpattern.hkls[0], xrdpattern.d_hkls[0], xrdpattern.x[0], xrdpattern.y[0]]
pic2 = [xrdpattern.hkls[1], xrdpattern.d_hkls[1], xrdpattern.x[1], xrdpattern.y[1]]
pic3 = [xrdpattern.hkls[2], xrdpattern.d_hkls[2], xrdpattern.x[2], xrdpattern.y[2]]

table = [["h,k,l", "d_hkl", r'$2\theta', 'Amplitude'], pic1, pic2, pic3]

display(HTML(tabulate.tabulate(table, tablefmt = 'html')))
```
